```python
data['EdLevel'] = data['EdLevel'].fillna('Unknown')
data['Country'] = data['Country'].fillna('Unknown')
data['Employment'] = data['Employment'].fillna('Unknown')
data.head(5)
```

                   Removing the Country which have less num of employee!

```python
def shorten_categories(categories, cutoff):
    categorical_map={}
    for i in range(len(categories)):
        if categories.values[i]>=cutoff:
            categorical_map[categories.index[i]]=categories.index[i]
        else:
            categorical_map[categories.index[i]]='others'
    return categorical_map
```

```python
country_map=shorten_categories(data.Country.value_counts(), 400)
data['Country']=data['Country'].map(country_map)
print(data.Country.value_counts())
```

```python
def clean_education(x):
    if 'Bachelor’s degree' in x:
        return 'Bachelor’s degree'
    if 'Master’s degree' in x:
        return 'Master’s degree'
    if 'Professional degree' in x or 'Something else' in x or 'Associate degree' in x:
        return 'Post grad'
    return 'Less than a Bachelors'

data['EdLevel'] = data['EdLevel'].apply(clean_education)
```

```python
data["EdLevel"].unique()
```

                                                Converting categorical data into numerical ones using LabelEncoder!

```python
from sklearn.preprocessing import LabelEncoder
le_education=LabelEncoder()
data['EdLevel']=le_education.fit_transform(data['EdLevel'])
data["EdLevel"].unique()
```

```python
le_country=LabelEncoder()
data['Country']=le_country.fit_transform(data['Country'])
data['Country'].unique()
```

                   Assigning x and y

```python
X=data.drop(['Salary'],axis=1)
y=data['Salary']
```

```python
from sklearn .linear_model import LinearRegression
linear_reg=LinearRegression()
linear_reg .fit(X,y.values)
```

```python
y_pred=linear_reg.predict(X)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np
error = np.sqrt(mean_squared_error(y, y_pred))
print("${:,.02f}".format(error))
```

```python
from sklearn.tree import DecisionTreeRegressor
dec_tree_reg = DecisionTreeRegressor(random_state=0)
dec_tree_reg.fit(X, y.values)
```

```python
y_pred = dec_tree_reg.predict(X)
```

```python
error2 = np.sqrt(mean_squared_error(y, y_pred))
print("${:,.02f}".format(error2))
```

```python
from sklearn.ensemble import RandomForestRegressor
random_forest_reg = RandomForestRegressor(random_state=0)
random_forest_reg.fit(X, y.values)
```

```python
y_pred = random_forest_reg.predict(X)
```

```python
error1 = np.sqrt(mean_squared_error(y, y_pred))
print("${:,.02f}".format(error1))
```

```python
import matplotlib.pyplot as plt

# Combine the three errors
model_errors = [error, error2, error1]
model_names = ['Linear Regression', 'Decision Tree', 'Random Forest']

# Plot the bar chart
plt.figure(figsize=(8, 5))
plt.bar(model_names, model_errors, color=['skyblue', 'lightgreen', 'salmon'])
plt.title('Model Comparison (Lower RMSE is Better)')
plt.ylabel('RMSE')
plt.xlabel('Models')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
from sklearn.model_selection import GridSearchCV

max_depth = [None, 2,4,6,8,10,12]
parameters = {"max_depth": max_depth}

regressor = DecisionTreeRegressor(random_state=0)
gs = GridSearchCV(regressor, parameters, scoring='neg_mean_squared_error')
gs.fit(X, y.values)
```

```python
regressor = gs.best_estimator_

regressor.fit(X, y.values)
y_pred = regressor.predict(X)
error3 = np.sqrt(mean_squared_error(y, y_pred))
print("${:,.02f}".format(error3))
```

```python
import matplotlib.pyplot as plt

# Step 1: Append the new tuned model's error
errors = [error, error2, error1, error3]  # Replace 2403.14 with variable if stored
model_names = ['Linear Regression', 'Decision Tree', 'Random Forest', 'Tuned decision tree']

# Step 2: Plot the updated bar chart
plt.figure(figsize=(8, 5))
plt.bar(model_names, errors, color='skyblue')
plt.title('Model Comparison (Lower RMSE is Better)')
plt.ylabel('RMSE in ₹')
plt.ylim(0, max(errors)+500)
for i, err in enumerate(errors):
    plt.text(i, err + 100, f"₹{err:,.2f}", ha='center')
plt.show()
```
